```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# SPLITING VALUES 
```

```python
X=df.drop(columns=['name','status'],axis=1)
```

```python
y=df['status']
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=3)
```

```python
scaler=StandardScaler() # USED TO UNDERSTAND NATURE OF DATA
```

```python
scaler.fit(X_train)
```

```python
X_train=scaler.transform(X_train)#Converts all values in same range

X_test=scaler.transform(X_test)
```

```python
print(X_train)
```

```python
#MODEL TRAINING
```

```python
#SUPPORT VECTOR MACHINE MODEL
```

```python
model=svm.SVC(kernel='linear') # GOING TO CLASSIFY INTO PARKINSONS POSITIVE AND NEGATIVE PATIENTS.
```

```python
model.fit(X_train,y_train)
```

```python
#MODEL EVALUATION
```

```python
#TRAINIGN DATA ACCURACY
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(y_train,X_train_prediction)
```

```python
training_data_accuracy

#89.70% accuracy
```

```python
# TEST DATA ACCURACY
```

```python
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(y_test,X_test_prediction)
```

```python
test_data_accuracy

#86.44% Accuracy
```
